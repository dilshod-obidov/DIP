```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a simple sine wave signal
t = np.linspace(0, 1, 500)  # Time vector (0 to 1 second, 500 samples)
frequency = 5  # Frequency of the sine wave (5 Hz)
signal = np.sin(2 * np.pi * frequency * t)  # Sine wave signal

# Compute the Fourier Transform of the signal
fft_result = np.fft.fft(signal)
frequencies = np.fft.fftfreq(len(signal), d=t[1] - t[0])

# Plot the original signal
plt.figure(figsize=(10, 5))

plt.subplot(2, 1, 1)
plt.plot(t, signal)
plt.title('Original Sine Wave')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')

# Plot the magnitude of the Fourier Transform
plt.subplot(2, 1, 2)
plt.plot(frequencies[:len(frequencies)//2], np.abs(fft_result)[:len(fft_result)//2])  # Plot positive frequencies
plt.title('Fourier Transform')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image in grayscale
image = cv2.imread('image.png', 0)

# Compute the 2D Fourier Transform of the image
f_transform = np.fft.fft2(image)
f_shifted = np.fft.fftshift(f_transform)  # Shift zero frequency component to the center
magnitude_spectrum = np.log(np.abs(f_shifted) + 1)  # Compute magnitude for visualization

# Display the original image and its magnitude spectrum
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(magnitude_spectrum, cmap='gray')
plt.title('Magnitude Spectrum')
plt.axis('off')

plt.show()
```
